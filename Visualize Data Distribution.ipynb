# Scatter plot

##### > การกระจาย (ส่วนใหญ่ plot แบบ 2 มิติ)

```python
irisdata = pandas.read_csv('C:\\Users\\HARUKA\\Downloads\\iris.data',header = None)
irisdata
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy() #ตัด Data มา 50 ตัวแรก
setosa_np[:5,:]
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60*setosa_np[:,2],c='r')
```

##### Color > r g b c m y k

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
      if i < 0.15:
            colors.append('b')
      elif i < 0.3:
            colors.append('g')
      elif i < 0.45:
            colors.append('y')
      else:
            colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60*setosa_np[:,2],c=colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60*setosa_np[:,2],c=setosa_np[:,3])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60*setosa_np[:,2],c=setosa_np[:,3],cmap='YlOrRd')
```

##### RGB

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60*setosa_np[:,2],c=np.array([[1,0.5,0.8]]))
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60*setosa_np[:,2],c=setosa_np[:,3],cmap='YlOrRd',marker='x')
```

## Visualize 2 data
##### > เพื่อดูว่า data 2 ตัวต่างกันอย่างไร

```python
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60*setosa_np[:,2],c=setosa_np[:,3],cmap='BuPu',marker='X')
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=60*versicolor_np[:,2],c=versicolor_np[:,3],cmap='BuPu',marker='P')
```

## Visualize 3 data
##### > เพื่อดูว่า data 3 ตัวต่างกันอย่างไร

```python
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60*setosa_np[:,2],c=setosa_np[:,3],cmap='BuPu',marker='o')
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=60*versicolor_np[:,2],c=versicolor_np[:,3],cmap='BuPu',marker='D')
plt.scatter(virginica_np[:,0],virginica_np[:,1],s=60*virginica_np[:,2],c=virginica_np[:,3],cmap='BuPu',marker='*')
```

#### Plot เทียบ 3 class หลายๆแบบ

### ex.

```python
plt.scatter(setosa_np[:,2],setosa_np[:,3],s=60*setosa_np[:,0],c=setosa_np[:,1],cmap='BuPu',marker='*')
plt.scatter(versicolor_np[:,2],versicolor_np[:,3],s=60*versicolor_np[:,0],c=versicolor_np[:,1],cmap='BuPu',marker='P')
plt.scatter(virginica_np[:,2],verginica_np[:,3],s=60*verginica_np[:,0],c=verginica_np[:,1],cmap='BuPu',marker='X')
```

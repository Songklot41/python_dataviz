# ข้อ 1
> สร้าง function ชื่อ smartDict ซึ่งทำหน้าที่:
- function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน โดย list แรกเป็น list ของตัวเลข list ที่สองเป็น list ของ text
- ถ้าค่าใน list แรก เป็นเลขคี่ ให้ค่า dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้าค่าใน list แรก เป็นเลขคู่ ให้ค่าใน dict  ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น
###### เช่น
            output1 = smartDict([1,2,3],['apple','banana','orange'])
            จะได้ output1 เป็น {1:'apple','banana':2,3:'orange'}
            
            output2 = smartDict([7,1,4,6],['A','B','E','Zzz'])
            จะได้ output2 เป็น {7:'A', 1:'B', 'E': 4, 'Zzz': 6}

```python
def smartDict (listnumber, listtext, logicV):
      if logicV: 
            retdict = {}
            for index in range(len(listnumber)):
                 retdict[listnumber[index]] = listtext[index]
                 
      else:
            retdict = {}
            for index in range(len(listtext)):
                 retdict[listtext[index]] = listnumber[index]
                    
      return retdict
```

```python
smartDict([1,2,3],['apple','banana','orange'],True)
```

```python
smartDict([7,1,4,6],['A','B','E','Zzz'],False )
```

##### กรดอะมิโนใด (A,T,G,C) มีปริมาณมากสุดและมีเท่าไหร่

```python
def countChar(seq,Character):
      charcount = 0
      for char in seq:
            if char == Character:
               charcount = charcount + 1 
      return charcount
```

```python
countChar('ATTAAAGGTTTATACCTTCCCAGGTAACAAACCAACCAACTTTCGATCTCTTGTAGATCTGTTCTCTAAA','A')
```

```python
ANum = 0
TNum = 0
GNum = 0
CNum = 0
for row in range (DNAseq.shape[0]):
      ANum += countChar(DNAseq.iloc[row][0],'A') 
      TNum += countChar(DNAseq.iloc[row][0],'T') 
      GNum += countChar(DNAseq.iloc[row][0],'G') 
      CNum += countChar(DNAseq.iloc[row][0],'C') 
        
print(f' There are {ANum} As, {TNum} Ts, {GNum} Gs, {CNum} Cs')
```

##### sequence ที่มีขนาด 3 และ 4 ที่เกิดซ้ำมากที่สุด

##### ขนาด 3 

```python
def populate3charstring(seq):
      str_list = list()
      for i in range(len(seq)-2):
           str_list.append(seq[i:i+3])
      return str_list
```

```python
populate3charstring('ATTAAAGGTT')
```

```python
list_all_3char = list()
for row in range(DNAseq.shape[0]):
      list_all_3char = list_all_3char + populate3charstring(DNAseq.iloc[row][0])
```

```python
len(list_all_3char)
```

```python
unique_3char = list(set(list_all_3char))
```

```python
len(unique_3char)
```

```python
dict_3char = {}
for sub_str in unique_3char:
      num = 0
      for s in list_all_3char:
            if sub_str == s:
                num += 1
      dict_3char[sub_str]   = num 
```

```python
dict_3char
```

```python
dict_3char['TAC']
```

```python
maxVal = -100
maxKey = None
for k in dict_3char.keys():
      if dict_3char[k] > maxVal:
            maxVal = dict_3char[k]
            maxKey = k
```

```python
maxKey
```

```python
print(f' max 3-long-sequence is {maxKey} with frequency equals {maxVal}')
```

##### สร้างตาราง Dictionary

```python
def splitstr(col1): 
    co1 = col1.split('/')
    return co1
```

```python
splitstr(df.iloc[0]['ภาษาอีสาน'])
```

```python
splitstr(df.iloc[1]['ภาษาไทย'])
```

```python
list1 = []
list2 = []
for i in range(len(Data)):
    if len(splitstr(Data.iloc[i]['ภาษาอีสาน'])) == len(splitstr(Data.iloc[i]['ภาษาไทย'])): 
        list1 += splitstr(Data.iloc[i]['ภาษาอีสาน']) 
        list2 += splitstr(Data.iloc[i]['ภาษาไทย'])
```

```python
def trans(col1,col2):
    str = f'{col1} แปลว่า {col2}'
    return str
```

```python
newnew = []
for i in range(len(list1)):
    if len(list1[i]) == len(list2[i]):
        newnew += [trans(list1[i],list2[i])]
```

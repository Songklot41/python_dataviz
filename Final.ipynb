```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
cov = pd.read_csv('C:\\Users\\HARUKA\\Desktop\\Aj.Ton\\python_dataviz\\2019_nCoV_data.csv') #เรียกข้อมูล
```

```python
cov.head()
```

```python
cov.shape
```

```python
cov.isnull().any() #ตรวจสอบค่า missing
```

```python
cov['Province/State']=cov['Province/State'].fillna('Other') #แก้ไขค่า missing
cov.isnull().any()
```

```python
cov1 = cov.groupby(['Country','Province/State']).max().reset_index() #จัดกลุ่ม(groupby)โดยใช้ค่า max
cov1
```

```python
cov2 = cov.groupby(['Country']).max().reset_index() #จัดกลุ่ม(groupby)โดยใช้ค่า max
cov2
```

```python
cov2.plot(kind='line', x='Country', y='Confirmed', color='Red')
plt.xlabel('Province/State',fontname='Tahoma')
plt.ylabel('Number of people infected',fontname='Tahoma')
plt.title('People inflected COVID-19 in each country ',fontname='Tahoma',fontsize=20)
plt.show()
```

```python
cov22 = cov2[cov2['Country']!='Mainland China']  #นำ Mainland China ออก 
cov22
```

### Number of people infected in each country

```python
cov22.plot(kind='bar', x='Country', y='Confirmed', color='Blue')
plt.xlabel('Country',fontname='Tahoma')
plt.ylabel('Number of people infected',fontname='Tahoma')
plt.title('People inflected COVID-19 in each country',fontname='Tahoma',fontsize=20)
plt.show()
```

### Number of people healed in each country

```python
cov22.plot(kind='bar', x='Country', y='Recovered', color='Green')
plt.xlabel('Country',fontname='Tahoma')
plt.ylabel('Number of people healed)',fontname='Tahoma')
plt.title('People healed in each country ',fontname='Tahoma',fontsize=20)
plt.show()
```

```python
x = cov22.groupby(['Country'], as_index=False).sum()['Country']
y1 = cov22.groupby(['Country'], as_index=False).sum()['Confirmed']
y2 = cov22.groupby(['Country'], as_index=False).sum()['Deaths']
y3 = cov22.groupby(['Country'], as_index=False).sum()['Recovered']
plt.figure(figsize=[10,5],dpi=60)
plt.gca(facecolor='yellow')
plt.plot(x,y1,'Blue',label = "Confirmed",marker='o')
plt.plot(x,y2,'Red',label = "Deaths",marker='o')
plt.plot(x,y3,'green',label = "Recovered",marker='o')
plt.legend()
plt.xticks(rotation = 90)
plt.xlabel('Country',fontname='Tahoma')
plt.ylabel('Numer',fontname='Tahoma')
plt.title('Confirmed,Deaths,Recovered in each country',fontname='Tahoma',fontsize=20)
```

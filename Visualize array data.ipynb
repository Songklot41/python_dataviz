```python
import pandas
```

```python
irisdata = pandas.read_csv('C:\\Users\\HARUKA\\Downloads\\iris.data',header = None)
irisdata
```

```python
import numpy as np
```

#จัดการตัวเลขใช้ .iloc 
#ตัวหน้า : คือแถว(ไม่ใส่อะไรเลยคือแถวแรก) ต่อมาเป็นหลัก(ไม่ใส่อะไรเลยคือแถวสุดท้าย) แถว,หลัก

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np
```

```python
versicolor_np =  irisdata.iloc[50:100,:-1].to_numpy()
virginica_np =  irisdata.iloc[100:,:-1].to_numpy()
```

## Visualize array

```python
from matplotlib import pyplot as plt
import matplotlib
```

#เพิ่มขนาดของ array
#เปลี่ยน array เป็นรูปภาพ

```python
matplotlib.rcParams['figure.figsize'] = [20,20]
plt.imshow(setosa_np,interpolation='nearest')
plt.show()
```

### plt.subplot (1,3,1) แถวที่ 1 หลักที่ 3 ลำดับที่ 1
#### > plot หลายๆรูปในรูปเดียว 

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.subplot(1,3,1)
plt.imshow(setosa_np, interpolation='nearest')
plt.subplot(1,3,2)
plt.imshow(versicolor_np, interpolation='nearest')
plt.subplot(1,3,3)
plt.imshow(virginica_np, interpolation='nearest')
plt.show()
```

### cmap = plt.get_cmap('Greys')
#### > การเปลี่ยนค่าสี 

```python
matplotlib.rcParams['figure.figsize'] = [20,20]
plt.imshow(setosa_np,interpolation='nearest',cmap = plt.get_cmap('Reds'))
plt.show()
```
